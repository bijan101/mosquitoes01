```python
im0 = mpimg.imread('image0.jpeg')
im1 = mpimg.imread('image1.jpeg')
im2 = mpimg.imread('image2.jpeg')
im3 = mpimg.imread('image3.jpeg')
im4 = mpimg.imread('image4.jpeg')

lower_black = np.array([0,0,0]  , dtype = "uint16")
upper_black = np.array([70,70,70], dtype = "uint16")
kernel      = np.ones((5,5),np.uint8)


im0 = cv2.inRange(im0, lower_black, upper_black)
im1 = cv2.inRange(im1, lower_black, upper_black)
im2 = cv2.inRange(im2, lower_black, upper_black)
im3 = cv2.inRange(im3, lower_black, upper_black)
im4 = cv2.inRange(im4, lower_black, upper_black)

e_im0 = cv2.bitwise_not(cv2.dilate(im0,kernel,iterations = 1))
e_im1 = cv2.bitwise_not(cv2.dilate(im1,kernel,iterations = 1))
e_im2 = cv2.bitwise_not(cv2.dilate(im2,kernel,iterations = 1))
e_im3 = cv2.bitwise_not(cv2.dilate(im3,kernel,iterations = 1))
e_im4 = cv2.bitwise_not(cv2.dilate(im4,kernel,iterations = 1))

plt.figure();plt.imshow(e_im0, cmap='gray')
plt.figure();plt.imshow(e_im1, cmap='gray')
plt.figure();plt.imshow(e_im2, cmap='gray')
plt.figure();plt.imshow(e_im3, cmap='gray')
plt.figure();plt.imshow(e_im4, cmap='gray')

```

```python
print(np.mean(np.where(e_im0<10),axis=1))
print(np.mean(np.where(e_im1<10),axis=1))
print(np.mean(np.where(e_im2<10),axis=1))
print(np.mean(np.where(e_im3<10),axis=1))
print(np.mean(np.where(e_im4<10),axis=1))
```
