## Scenario 1: Complete Space Randomness

```python
import numpy as np
import matplotlib.pyplot as plt

n_samples = 10000
x_bias = .01
y_bias = -.3
x_orig = 500*np.random.random(n_samples); x = x_orig
y_orig = 500*np.random.random(n_samples); y = y_orig
x_n = x+x_bias
y_n = y+y_bias
plt.figure(figsize=(20,5))
plt.scatter(x  , y  ,s=10)
plt.scatter(x_n, y_n,s=10)
plt.show()

x_n1 =[]
y_n1 =[]
for i in range(n_samples):
    x_min = []
    for j in range(x.shape[0]):
        x_min.append((x_n[j] - x[0])**2 + (y_n[j] - y[0])**2)
    x = np.delete(x,0)
    y = np.delete(y,0)
    x_n1.append(x_n[np.argmin(x_min)])
    y_n1.append(y_n[np.argmin(x_min)])
    x_n = np.delete(x_n,np.argmin(x_min))   
    y_n = np.delete(y_n,np.argmin(x_min))

x_dist = np.mean(x_n1-x_orig);print(f"x_bias = {x_bias}, x_dist = {x_dist}")
y_dist = np.mean(y_n1-y_orig);print(f"y_bias = {y_bias}, y_dist = {y_dist}")
```
